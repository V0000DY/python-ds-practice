# Read .csv from Multiple URLs with Pandas

https://www.football-data.co.uk/mmz4281/2122/E0.csv

Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

# Multiple leagues

```python
# creating list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames = []

# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2122' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element
frames[4]
```

# Multiple Seasons

```python
# creating list of leagues
leagues = ['E0','E2', 'E3']
frames = []

# loop through range of seasons and insert new column
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + league + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
frames[10]
```

```python
# total frames
len(frames)
```

```python
# show 1st, and last element
frames[17]
```

# Organize all the data in dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {'Spanish La Liga': 'SP1', 'Spanish Segunda Division': 'SP2', 'German Bundesliga': 'D1', 'English Premier League': 'E0', 'English League 1': 'E2', 'English League 2': 'E3'}
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

# creating list of leagues

# loop through range of seasons and insert new column
for league in dict_countries:
    frames = []
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['English Premier League']
```
