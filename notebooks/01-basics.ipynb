# If Statement

if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>

```python
age = 20

if age >= 18:
    print('You are an adult!')
elif age >= 13:
    print('You are a teenager!')
else:
    print('You are a kid!')
```

# For loop

for <variable> in <iterable>:
    <code>

```python
countries
```

```python
for i, country in enumerate(countries):
    print(i)
    print(country)
```

```python
my_data = {'name':'Frank', 'age':26}
my_data
```

```python
for key, value in my_data.items():
    print(key)
    print(value)
```
